# Ollama Installation

## Step 1: Install Ollama
- Install Ollama from [here](https://ollama.com/)
- Run the executable (OllamaSetup.exe)

- Some antiviruses might block, so temporarily disabling or ignoring is advised
- Install the appropriate model
  - `ollama pull <model>`
  - If not already installed, it will automatically install
  - [List of models](https://ollama.com/library)
  - Used mistral

## Step 2: Install Ollama-privateGPT
- [Github repo](https://github.com/PromptEngineer48/Ollama/tree/main)
- Recommended to create a virtual environment
- Used anaconda
- In the anaconda prompt, enter `conda create -n privategpt python=3.11`
  - Python 3.11 works, but some earlier version do not work
  - If encountering issues installing libraries, likely due to outdated python version
- Activate virtual env `activate privategpt`
- Clone the repo `git clone https://github.com/PromptEngineer48/Ollama.git`
- Enter the folder "2-ollama-privateGPT-chat-with-docs"
- Install the requirements `pip install -r requirements.txt`
  - If errors occur, make sure python version is up to date

## Step 3: Activate Ollama-privateGPT
- If not already done, pull the model
  - `ollama pull <model>`
- Make new directory for all documents
  - `mkdir source_docs`
- Put the documents into the new folder
- "Ingest" all the documents (have them processed)
  - `python ingest.py`
- Once ingestion is complete, run the privateGPT
  - `python privateGPT.py`
